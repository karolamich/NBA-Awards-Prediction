```python
df_team_allnba.rename(columns={'W_PCT': 'W_TEAM'}, inplace=True)
df_team_rookie.rename(columns={'W_PCT': 'W_TEAM'}, inplace=True)
```

```python
df_teams_allnba_cut = df_team_allnba[['TEAM_ID', 'SEASON', 'W_TEAM']]
df_teams_rookie_cut = df_team_rookie[['TEAM_ID', 'SEASON', 'W_TEAM']]
```

```python
df_basic_teams_allnba = pd.merge(df_basic_allnba, df_teams_allnba_cut, on=['TEAM_ID', 'SEASON'], how='left')
df_basic_teams_rookie = pd.merge(df_basic_allnba, df_teams_rookie_cut, on=['TEAM_ID', 'SEASON'], how='left')
```

```python
df_basic_teams_allnba.info()
```

```python
df_merge_allnba = pd.merge(df_basic_teams_allnba, df_adv_allnba, on=['PLAYER_ID', 'SEASON'])
df_merge_rookie = pd.merge(df_basic_teams_rookie, df_adv_rookie, on=['PLAYER_ID', 'SEASON'])
```

```python
text_columns = df_merge_allnba.select_dtypes(include=['object']).columns
print(text_columns)
text_columns_rookie = df_merge_rookie.select_dtypes(include=['object']).columns
print(text_columns_rookie)
```

```python
from unidecode import unidecode

df_merge_allnba['PLAYER_NAME'] = df_merge_allnba['PLAYER_NAME'].apply(unidecode).str.strip()
df_voting_allnba['PLAYER_NAME'] = df_voting_allnba['PLAYER_NAME'].apply(unidecode).str.strip()

df_merge_rookie['PLAYER_NAME'] = df_merge_rookie['PLAYER_NAME'].apply(unidecode).str.strip()
df_voting_rookie['PLAYER_NAME'] = df_voting_rookie['PLAYER_NAME'].apply(unidecode).str.strip()
```

```python
brak_dopasowania = df_voting_allnba[
    ~df_voting_allnba['PLAYER_NAME'].isin(df_merge_allnba['PLAYER_NAME'])
]
print(brak_dopasowania[['PLAYER_NAME', 'SEASON']])
```

```python
name_stats = df_merge_allnba[df_merge_allnba['PLAYER_NAME'].str.contains('Butler')]['PLAYER_NAME'].unique()
name_voting = df_voting_allnba[df_voting_allnba['PLAYER_NAME'].str.contains('Butler')]['PLAYER_NAME'].unique()

print(name_stats)
print(name_voting)
```

```python
df_merge_allnba['PLAYER_NAME'] = df_merge_allnba['PLAYER_NAME'].replace(
    'Jimmy Butler III', 'Jimmy Butler'
)
```

```python
brak_dopasowania = df_voting_allnba[
    ~df_voting_allnba['PLAYER_NAME'].isin(df_merge_allnba['PLAYER_NAME'])
]
print(brak_dopasowania[['PLAYER_NAME', 'SEASON']])
```

```python
brak_dopasowania_rookie = df_voting_rookie[
    ~df_voting_rookie['PLAYER_NAME'].isin(df_merge_rookie['PLAYER_NAME'])
]
print(brak_dopasowania_rookie[['PLAYER_NAME', 'SEASON']])
```

```python
name_stats_rookie = df_merge_rookie[df_merge_rookie['PLAYER_NAME'].str.contains('Jackson')]['PLAYER_NAME'].unique()
name_voting_rookie = df_voting_rookie[df_voting_rookie['PLAYER_NAME'].str.contains('Jackson')]['PLAYER_NAME'].unique()

print(name_stats_rookie)
print(name_voting_rookie)
```

```python
df_voting_rookie['PLAYER_NAME'] = df_voting_rookie['PLAYER_NAME'].replace(
    'GG Jackson II', 'GG Jackson'
)
```

```python
brak_dopasowania_rookie = df_voting_rookie[
    ~df_voting_rookie['PLAYER_NAME'].isin(df_merge_rookie['PLAYER_NAME'])
]
print(brak_dopasowania_rookie[['PLAYER_NAME', 'SEASON']])
```

```python
df_stats_voting_allnba = pd.merge(df_merge_allnba, df_voting_allnba, on=['PLAYER_NAME', 'SEASON'], how='left')
df_stats_voting_rookie = pd.merge(df_merge_rookie, df_voting_rookie, on=['PLAYER_NAME', 'SEASON'], how='left')
```

```python
df_stats_voting_allnba.info()
df_stats_voting_rookie.info()
```

```python
df_stats_voting_allnba['PTS_WON'] = df_stats_voting_allnba['PTS_WON'].fillna(0)
df_stats_voting_allnba['PTS_MAX'] = df_stats_voting_allnba['PTS_MAX'].fillna(1)

df_stats_voting_rookie['PTS_WON'] = df_stats_voting_rookie['PTS_WON'].fillna(0)
df_stats_voting_rookie['PTS_MAX'] = df_stats_voting_rookie['PTS_MAX'].fillna(1)
```

```python
df_stats_voting_allnba['TARGET'] = df_stats_voting_allnba['PTS_WON'] / df_stats_voting_allnba['PTS_MAX']
df_stats_voting_rookie['TARGET'] = df_stats_voting_rookie['PTS_WON'] / df_stats_voting_rookie['PTS_MAX']
```
